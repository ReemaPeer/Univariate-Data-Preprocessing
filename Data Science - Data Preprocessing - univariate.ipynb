# Filling the null values automatically using Imputer

```python
dataset = pd.read_csv('placement.csv')
dataset
```

```python
# Splitting the qualitative and quantitative data

def QuanQual(dataset):
    
    Qual = []
    Quan = []

    for col in dataset.columns:
        if dataset[col].dtype == 'O':
            Qual.append(col)
        else:
            Quan.append(col)
    return Quan,Qual         
```

```python
Quan,Qual = QuanQual(dataset)
```

# InterQuartile Range
      Formulas:
         IQR = Q3 -Q1
         Lesser Outlier = Q1 - (1.5)*IQR
         Greater Outlier = Q3 + (1.5)*IQR

```python
# Creating the table for data description manually

description = pd.DataFrame(index = ['Mean', 'Median', 'Mode', 'Q1-25%', 'Q2-50%', 'Q3-75%', '99%', 'Q4-100%', 'IQR', 
                                   'Lesser', 'Greater', '1.5Range', 'Min', 'Max'], columns = Quan)
description 
```

```python
# Filling the descriptive table with respective values

for col in Quan:
    description[col]['Mean'] = data[col].mean()
    description[col]['Median'] = data[col].median()
    description[col]['Mode'] = data[col].mode()[0]
    description[col]['Q1-25%'] = np.percentile(data[col],25)
    description[col]['Q2-50%'] = np.percentile(data[col],50)
    description[col]['99%'] = np.percentile(data[col],99)
    description[col]['Q3-75%'] = np.percentile(data[col],75)
    description[col]['Q4-100%'] = np.percentile(data[col],100)
    description[col]['IQR'] = description[col]['Q3-75%'] - description[col]['Q1-25%']
    description[col]['1.5Range'] = 1.5*description[col]['IQR']
    description[col]['Lesser'] = description[col]['Q1-25%'] - description[col]['1.5Range']
    description[col]['Greater'] = description[col]['Q3-75%'] + description[col]['1.5Range']
    description[col]['Min'] = data[col].min()
    description[col]['Max'] = data[col].max()
```

```python
description
```

```python
# Creating a function for probability density function

def get_pdf(data,startrange,endrange):
    
    import matplotlib.pyplot as plt
    import seaborn as sns
    from scipy.stats import norm
    
    plot = sns.distplot(data, kde = True, kde_kws = {'color' : 'blue'}, color = 'green')
    plt.axvline(startrange, color = 'red')
    plt.axvline(endrange, color = 'red')
    
    # Generating the sample data
    
    sample = data
    
    # Calculating the parameters
    
    sample_mean = sample.mean()
    sample_std = sample.std()
    print('\t Mean = %.3f \n Standard Deviation = %.3f '% (sample_mean,sample_std))
    
    # Define the distribution
    
    dist = norm(sample_mean, sample_std)
    
    # Sample probability for the given range of outcomes
    
    probabilities = []
    for value in range(startrange, endrange):
        probability = dist.pdf(value)
        probabilities.append(probability)
    prob = sum(probabilities)
    print('The area between the given range {} to {} is: {}'.format (startrange, endrange, prob))
    return prob
    
    
```

# Standard Normal Distribution

```python
# Converting to standard normal distribution


Mean = data['hsc_p'].mean()
Std = data['hsc_p'].std()

z_scores = []
for i in data['hsc_p']:
    z_score = (i - Mean) / Std
    z_scores.append(z_score)

# Plotting the distribution of z-scores
sns.distplot(z_scores, kde=True)

# Calculating the average of the z-scores
average_z_score = sum(z_scores) / len(z_scores)
average_z_score
```

```python
# Creating a function for Converting to standard normal distribution

def stdND_graph(data):
    import seaborn as sns
    Mean = data.mean()
    Std = data.std()
    
    z_scores = []
    for values in data:
        z_score = (values-Mean)/Std
        z_scores.append(z_score)
        
# Plotting the distribution of z_scores

    sns.distplot (z_scores, kde = True)

# Calculating the average of z_scores

    average_z_scores = sum(z_scores)/len(z_scores)
    return average_z_scores
    
```

```python
stdND_graph(data['hsc_p'])
```
